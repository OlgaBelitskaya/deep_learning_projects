## Step 2. Preprocess

```python
hide_code
# Create vocabulary
def create_lookup_tables(text):
    vocabulary = set(text)
    int_to_vocab = dict(enumerate(vocabulary))
    vocab_to_int = dict((v,k) for k,v in int_to_vocab.items())
    
    return vocab_to_int, int_to_vocab
```

```python
hide_code
# Russian
vocab_to_int_russian, int_to_vocab_russian = create_lookup_tables(words_russian)
int_text_russian = [vocab_to_int_russian[word] for word in words_russian]
print(int_text_russian[:7])
print(words_russian[:7])
```

```python
hide_code
# Russian
part_to_int_russian, int_to_part_russian = create_lookup_tables(parts_of_speech_russian)
int_parts_russian = [part_to_int_russian[part] for part in parts_of_speech_russian]
print(int_parts_russian[:7])
print(parts_of_speech_russian[:7])
```

```python
hide_code
# Physics
vocab_to_int_physics, int_to_vocab_physics = create_lookup_tables(words_physics)
int_text_physics = [vocab_to_int_physics[word] for word in words_physics]
print(int_text_physics[:7])
print(words_physics[:7])
```

```python
hide_code
# Physics
part_to_int_physics, int_to_part_physics = create_lookup_tables(parts_of_speech_physics)
int_parts_physics = [part_to_int_physics[part] for part in parts_of_speech_physics]
print(int_parts_physics[:7])
print(parts_of_speech_physics[:7])
```

## Step 3. Create Artificial Variables
#### Sentence Structures

```python
hide_code
set(df_russian['parts_of_speech'])
```

```python
hide_code
set(df_physics['parts_of_speech'])
```

```python
hide_code
n_russian = int(len(df_russian['parts_of_speech'])/7)
sentence_structures_russian = df_russian['parts_of_speech'].values.reshape(n_russian,7)
sentence_structures_russian.shape
```

```python
hide_code
n_physics = int(len(df_physics['parts_of_speech'])/7)
sentence_structures_physics = df_physics['parts_of_speech'].values.reshape(n_physics,7)
sentence_structures_physics.shape
```

```python
hide_code
sentence_structures = numpy.concatenate((sentence_structures_russian,
                                         sentence_structures_physics),
                                        axis=0)
parts_of_speech_set, sentence_structures = numpy.unique(sentence_structures, return_inverse=True)
sentence_structures = to_categorical(sentence_structures, 11)
sentence_structures = sentence_structures.reshape(int((n_russian+n_physics)),7, 11)
sentence_structures.shape
```

#### Word Lengths

```python
hide_code
set(df_russian['word_lengths'])
```

```python
hide_code
set(df_physics['word_lengths'])
```

```python
hide_code
sentence_lengths_russian = df_russian['word_lengths'].values.reshape(n_russian,7)
sentence_lengths_russian.shape
```

```python
hide_code
sentence_lengths_physics = df_physics['word_lengths'].values.reshape(n_physics,7)
sentence_lengths_physics.shape
```

```python
hide_code
sentence_lengths = numpy.concatenate((sentence_lengths_russian,
                                         sentence_lengths_physics),
                                        axis=0)
length_set, sentence_lengths = numpy.unique(sentence_lengths, return_inverse=True)
sentence_lengths = to_categorical(sentence_lengths-1, 19)
sentence_lengths = sentence_lengths.reshape(int((n_russian+n_physics)),7,19)
sentence_lengths.shape
```

#### Features and Targets

```python
hide_code
x = numpy.concatenate((sentence_structures.reshape((n_russian+n_physics)*7,11),
                       sentence_lengths.reshape((n_russian+n_physics)*7,19)),
                      axis=1)
x = x.reshape(n_russian+n_physics,7*(11+19))
x.shape
```

```python
hide_code
zeros = numpy.zeros((n_russian), dtype=int)
ones = numpy.ones((n_physics), dtype=int)
y = numpy.concatenate((zeros, ones), axis=0)
y.shape
```

```python
x_train, x_test, y_train, y_test = \
train_test_split(x, y, test_size=0.3, random_state=1)

n = int(len(x_test)/2)
x_valid, y_valid = x_test[:n], y_test[:n]
x_test, y_test = x_test[n:], y_test[n:]

x_train.shape, x_test.shape, x_valid.shape, \
y_train.shape, y_test.shape, y_valid.shape
```

## Step 4. Build the Models
#### MLP

```python
hide_code
def mlp_model():
    model = Sequential()

    model.add(Dense(1024, activation='relu', 
                    input_shape = x_train.shape[1:]))
    model.add(Dropout(0.1)) 
    
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.1))
    
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='nadam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
hide_code
mlp_model = mlp_model()
mlp_checkpointer = ModelCheckpoint(filepath='weights.best.sentence.mlp.hdf5', 
                                   verbose=2, save_best_only=True)
mlp_model.summary()
```

```python
hide_code
mlp_history = mlp_model.fit(x_train, y_train, 
                            validation_data=(x_valid, y_valid), 
                            epochs=5, batch_size=8, verbose=2,
                            callbacks=[mlp_checkpointer]);
```

```python
hide_code
mlp_model.load_weights('weights.best.sentence.mlp.hdf5')
mlp_score = mlp_model.evaluate(x_test, y_test)
mlp_score
```

#### CNN

```python
hide_code
x_train = x_train.reshape(-1,7,30)
x_test = x_test.reshape(-1,7,30) 
x_valid = x_valid.reshape(-1,7,30)
```

```python
hide_code
def cnn_model():
    model = Sequential()
    model.add(Conv1D(32, 3, padding='same', activation='relu',
                     input_shape=(7,30)))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.25))

    model.add(Conv1D(96, 3, padding='same', activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.25))

    model.add(Flatten())
    
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
hide_code
cnn_model = cnn_model()
cnn_checkpointer = ModelCheckpoint(filepath='weights.best.sentence.cnn.hdf5', 
                                   verbose=2, save_best_only=True)
cnn_model.summary()
```

```python
hide_code
cnn_history = cnn_model.fit(x_train, y_train, 
                            validation_data=(x_valid, y_valid), 
                            epochs=15, batch_size=8, verbose=2,
                            callbacks=[cnn_checkpointer]);
```

```python
hide_code
cnn_model.load_weights('weights.best.sentence.cnn.hdf5')
cnn_score = cnn_model.evaluate(x_test, y_test)
cnn_score
```

#### RNN

```python
hide_code
x_train = x_train.reshape(-1,1,210)
x_test = x_test.reshape(-1,1,210) 
x_valid = x_valid.reshape(-1,1,210)
```

```python
hide_code
def rnn_model():
    model = Sequential()

    model.add(LSTM(210, return_sequences=True, input_shape=(1, 210)))    
    model.add(LSTM(210, return_sequences=True))
    
    model.add(LSTM(512))  
    
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy', optimizer='nadam', metrics=['accuracy'])    
    return model 
```

```python
hide_code
rnn_model = rnn_model()
rnn_checkpointer = ModelCheckpoint(filepath='weights.best.sentence.rnn.hdf5', 
                                   verbose=2, save_best_only=True)
rnn_model.summary()
```

```python
hide_code
rnn_history = rnn_model.fit(x_train, y_train, 
                            validation_data=(x_valid, y_valid), 
                            epochs=5, batch_size=8, verbose=2,
                            callbacks=[rnn_checkpointer]);
```

```python
hide_code
rnn_model.load_weights('weights.best.sentence.rnn.hdf5')
rnn_score = rnn_model.evaluate(x_test, y_test)
rnn_score
```
